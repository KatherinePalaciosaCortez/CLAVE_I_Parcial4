```python
#  Cargar el archivo CSV

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/KatherinePalaciosaCortez/CLAVE_I_Parcial4/refs/heads/main/data/raw/clave_I_asociacion.csv")
```

```python
#valores duplicados
print("\nCantidad de registros duplicados:")
print(df.duplicated().sum())
```

```python

# tipos de datos
print("\nInformación general del dataset:")
print(df.info())
```

```python
# 4. Selección de columnas de asociación
columnas_asociacion = [
    "categoria", "item", "canal"
    ]
df_asociacion = df[columnas_asociacion].copy()
df_asociacion_bool = df_asociacion.astype(bool)

# 5. Patrones frecuentes y reglas
reglas = []
n = len(df_asociacion_bool)

for k in range(2, len(columnas_asociacion) + 1):
    for itemset in combinations(columnas_asociacion, k):
        itemset = set(itemset)
        soporte_itemset = df_asociacion_bool[list(itemset)].all(axis=1).mean()
        if soporte_itemset < 0.10:
            continue

        for tam_antecedente in range(1, k):
            for antecedente in combinations(itemset, tam_antecedente):
                antecedente = set(antecedente)
                consecuente = itemset - antecedente

                soporte_antecedente = df_asociacion_bool[list(antecedente)].all(axis=1).mean()
                soporte_consecuente = df_asociacion_bool[list(consecuente)].all(axis=1).mean()
                confianza = soporte_itemset / soporte_antecedente
                lift = confianza / soporte_consecuente

                if confianza >= 0.50:
                    reglas.append({
                        "antecedente": ", ".join(sorted(antecedente)),
                        "consecuente": ", ".join(sorted(consecuente)),
                        "soporte": soporte_itemset,
                        "confianza": confianza,
                        "lift": lift
                    })

reglas_df = pd.DataFrame(reglas).sort_values(
    by=["confianza", "soporte"], ascending=False
)
print(reglas_df)
```
