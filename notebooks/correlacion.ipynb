```python
import pandas as pd
import numpy as np

# 1. Cargar archivo
df = pd.read_csv("https://raw.githubusercontent.com/KatherinePalaciosaCortez/CLAVE_I_Parcial4/refs/heads/main/data/raw/clave_I_correlacion.csv")


```

```python
# 2. Revision inicial
print("Dimensiones del dataset:")
print(df.shape)

print("Primeras filas:")
print(df.head())

print("Informacion general:")
print(df.info())

```

```python

# 3. Revision de valores nulos
print("Valores nulos:")
print(df.isnull().sum())



```

```python

# 4. Estadistica descriptiva
print("Estadistica descriptiva:")
print(df.describe())
```

```python
# 5. Revision de duplicados
print("Registros duplicados:")
print(df.duplicated().sum())
```

```python
# 6. Deteccion de valores atipicos
variables_numericas = [
    "edad",
    "ingresos_mensuales",
    "frecuencia_compra",
    "ticket_promedio",
    "reclamos",
    "uso_app",
    "tiempo_respuesta",
 "consumo_total"
    ]

print("evision de valores atipicos:")
for columna in variables_numericas:
    media = df[columna].mean()
    desviacion = df[columna].std()

    limite_inferior = media - 2 * desviacion
    limite_superior = media + 2 * desviacion

    atipicos = df[
        (df[columna] < limite_inferior) |
        (df[columna] > limite_superior)
    ]

    print("Variable:", columna)
    print("Media:", round(media, 2))
    print("Desviacion estandar:", round(desviacion, 2))
    print("Limite inferior:", round(limite_inferior, 2))
    print("Limite superior:", round(limite_superior, 2))
    print("Cantidad de valores atipicos:", len(atipicos))
    print("-" * 40)

```

```python
import numpy as np
# 7. Matriz de correlacion

# Excluir la columna 'cliente_id' ya que no es numérica y no se puede correlacionar
matriz_correlacion = df.drop(columns=['cliente_id']).corr()

print("Matriz de correlacion:")
print(matriz_correlacion.round(3))
```

```python
# 8. Correlaciones con la variable principal consumo_total
print("Correlaciones con frecuencia_compra:")
print(matriz_correlacion["frecuencia_compra"].sort_values(ascending=False).round(3))

```
